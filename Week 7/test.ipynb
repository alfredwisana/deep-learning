```python
data = cv2.imread('ant.jpg')
cropped_image = data[100:420, 100:420] 
gray = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
hist = cv2.calcHist(gray,channels=[0], mask=None, histSize=[32], ranges=256)
plt.imshow(hist)
```
