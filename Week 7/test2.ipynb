```python
import cv2
import matplotlib.pyplot as plt
```

```python
data = cv2.imread("ant.jpg")
```

```python
crop = data[150:800, 350:1000]
```

```python
plt.imshow(data)
```

```python
plt.imshow(crop)
```

```python
data.shape
```

```python
result = data
result[:,:,0] = cv2.equalizeHist(data[:,:,0])
result[:,:,1] = cv2.equalizeHist(data[:,:,1])
result[:,:,2] = cv2.equalizeHist(data[:,:,2])

histg = cv2.calcHist([result],[0],None,[256],[0,256])
plt.plot(histg)
histg = cv2.calcHist([result],[1],None,[256],[0,256])
plt.plot(histg)
histg = cv2.calcHist([result],[2],None,[256],[0,256])
plt.plot(histg)
```

```python
plt.imshow(result)
```

```python
grayImg = cv2.cvtColor(data, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImg)
```

```python
histg = cv2.calcHist([grayImg],[0],None,[256],[0,256])
plt.plot(histg)
```

```python
result = cv2.equalizeHist(grayImg)
plt.imshow(result)
```

```python
histg = cv2.calcHist([result],[0],None,[256],[0,256])
plt.plot(histg)
```
